# **★ Job Title ★**

```python
Ai2.Job_Title.value_counts()
```

```python
value_counts = Ai2['Job_Title'].value_counts().reset_index()
value_counts.columns = ['Job_Title', 'Count']
# Create a bar chart
fig = px.bar(
    value_counts,
    x='Job_Title',
    y='Count',
    title='Count of Each Job Title',
    labels={'Count': 'Count', 'Job_Title': 'Job Title'},
    color='Count',
    color_continuous_scale='Viridis'
)
# Update layout for better visualization
fig.update_layout(
    xaxis_title='Job Title',
    yaxis_title='Count',
    title_x=0.5,
    xaxis=dict(
        tickangle=-45
    )
)

fig.show()
```

```python
# Conclusion:-
'''
--> The code visualizes the distribution of different job titles within the dataset using a bar chart.
     This allows for easy comparison of the frequencies of various job titles, providing insights
     into the prevalence of each role in the AI job market.

--> The use of color further enhances the visualization by representing the count of each job title.

'''
```
